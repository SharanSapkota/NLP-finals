```python
import task8_news as t8
import pandas as pd

# Keywords to analyze
keywords = ["nature", "resilience", "sustainability", "climate change"]

# Step 0: define time periods (last 1 month due to free NewsAPI limit)
periods = t8.get_time_periods(months=1)  # only 1 month for free plan

# To store all periods together if needed
all_data = []

for start, end in periods:
    period_label = f"{start}_to_{end}"
    print(f"Processing period: {period_label}")

    # Step 1: Fetch news for this period
    df_period = t8.fetch_news(keywords, (start, end))
    if df_period.empty:
        print(f"No news found for period {period_label}, skipping further steps.")
        continue

    df_period.to_csv(f"news_dataset_{period_label}.csv", index=False)

    # Step 2: Generate WordClouds for this period
    t8.generate_wordcloud(df_period, period_label)

    # Step 3: Compute TF-IDF similarity for this period
    tfidf_sim = t8.compute_tfidf_similarity(df_period)
    tfidf_sim.to_csv(f"tfidf_similarity_{period_label}.csv")

    # Step 4: Compute Word2Vec similarity for this period
    w2v_sim = t8.compute_w2v_similarity(df_period)
    w2v_sim.to_csv(f"w2v_similarity_{period_label}.csv")

    all_data.append(df_period)

if all_data:
    df_all = pd.concat(all_data, ignore_index=True)
    df_all.to_csv("news_dataset_all_periods.csv", index=False)
```
